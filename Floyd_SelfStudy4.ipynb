```
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```
movies.head()
```

```
movies.shape
```

```
#Examine only movies that are a specfic length (filtering)
```

```
type(True)
```

```
type(False)
```

```
booleans = []  #long way to do it
for length in movies.duration:
  if length >= 200:
    booleans.append(True)
  else:
    booleans.append(False)
```

```
booleans[0:5]
```

```
len(booleans)
```

```
is_long = pd.Series(booleans) #converts booleans to a panda series
```

```
is_long.head()
```

```
movies[is_long] # only show those that are true
```

```
is_long = movies.duration >= 200
```

```
is_long.head()
```

```
movies[is_long]
```

```
movies[movies.duration >= 200] # another quick way to do it
```

```
#Bonus Tip: how do we just get the genre?
```

```
movies[movies.duration >= 200].genre #dot notation way
```

```
movies[movies.duration >= 200]['genre'] #bracket wasy
```

```
movies.loc[movies.duration >= 200]['genre'] # select by labels using loc. best practice way when using a condition
```

```
#Logical operators and and or
```

```
True or False # results is True, if either is true then the whole thing is true
```

```
False or False
```

```
True or True
```

```
#and works only if both are true
```

```
True and False
```

```
movies[(movies.duration >= 200) & (movies.genre =='Drama')] # and condition but use &
```

```
(movies.duration >= 200) & (movies.genre =='Drama')
```

```
#Video 14 - When should I use a 'groupby'  in pandas
```

```
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```
drinks.head()
```

```
drinks.beer_servings.mean()
```

```
# Lets do it by continent using group by
```

```
drinks.groupby('continent').beer_servings.mean()
```

```
drinks[drinks.continent=='Africa']
```

```
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```
drinks[drinks.continent=='Europe'].beer_servings.mean()
```

```
drinks[drinks.continent=='North America'].beer_servings.mean()
```

```
drinks[drinks.continent=='Africa'].beer_servings.max()
```

```
drinks.groupby('continent').beer_servings.max()
```

```
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```
drinks.groupby('continent').mean() # Bonus tip
```

```
# Video 15 - How do I explore a pandas Series
```

```
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```
movies.head()
```

```
movies.dtypes # data types
```

```
movies.genre.describe() #Count, unique values, most/top, freq
```

```
movies.genre.value_counts() # how much it appears in this series
```

```
movies.genre.value_counts(normalize=True) # changes to percentage
```

```
type(movies.genre.value_counts())
```

```
movies.genre.value_counts().head() # since the type is a series you can use head on it
```

```
movies.genre.unique() # you can see all the unique values in this series
```

```
movies.genre.nunique()  # you can see number of unique values in this series
```

```
pd.crosstab(movies.genre, movies.content_rating) #tally of how many belongs to each combination
```

```
movies.duration.describe() # describe works on numeric values too
```

```
movies.duration.mean()
```

```
movies.duration.value_counts() # movies with their duration as a count
```
